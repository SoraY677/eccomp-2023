```python
import glob

files = glob.glob("./s0/rs0-0000000*-2-*.json")
```

```python
import json

objective_transition = []
for file in files:
    with open(file) as f:
        item = json.load(f)
        objective_list = []
        for evalution in item['evaluation_list']:
            objective_list.append(evalution['eval']['objective'])
        objective_transition.append(sum(objective_list) / len(objective_list))
print(objective_transition)
        
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 折れ線グラフを出力
left = np.array([i + 1 for i in range(len(objective_transition))])
height = np.array(objective_transition)
plt.plot(left, height)
plt.xlabel("count")
plt.ylabel("objective average")
```
